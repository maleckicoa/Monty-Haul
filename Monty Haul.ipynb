#### Viz Function

```python
def viz_func(win_tracker):
    imag=[]
    goat='/home/aleksa/Desktop/Goats and Cars/goat.png'
    money='/home/aleksa/Desktop/Goats and Cars/cars.png'

    for i in range(0,60): 
        if win_tracker[i]==0:
            imag.append(goat)
        else:
            imag.append(money)

    %matplotlib inline

    images = []
    for img_path in imag:
        images.append(mpimg.imread(img_path))

    plt.figure(figsize=(40,10))# 
    columns = 20
    for i, image in enumerate(images):  
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        
        plt.imshow(image)
        plt.axis('off')
               
```

#### No Switch

```python
prize_list = ['Goat', 'Goat', 'Car']                   # the 3 doors hiding 2 goats and a car
choice_list = [1,2,3]                                  # the 3 choices we have


choice_list = [x-1 for x in choice_list]
win_tracker_no_change=[]

for i in range(1,100000):
    random.shuffle(prize_list)                          # shuffle the goats and a car
    my_choice=random.choice(choice_list)                # randomly choose a door
    if 'Car'== prize_list[random.choice(choice_list)]:  # check if we won
        win_tracker_no_change.append(1)                 # add 1 if we won
    else:
        win_tracker_no_change.append(0)                 # add 0 if we lost

print('\n\n', 'WINNING PROBABILITY: ', sum(win_tracker_no_change)/len(win_tracker_no_change), '\n\n')

viz_func(win_tracker_no_change)
```

#### Switch

```python
prize_list = ['Goat', 'Goat', 'Car']       # the 3 doors hiding 2 goats and a car
choice_list = [1,2,3]                      # the 3 choices we have

choice_list = [x-1 for x in choice_list]
win_tracker_yes_change=[]

for i in range(1,100000):
    random.shuffle(prize_list)              # shuffle the goats and a car
    my_choice=random.choice(choice_list)    # randomly choose a door     
    ml= prize_list.copy()    
    del ml[my_choice]                       # eliminate one door hiding a goat and eliminate our first choice (this is the same as swithing)
    if ml[0]=="Goat":
        del ml[0]
    else:
        del ml[1]        
    if 'Car'== ml[0]:                       # check if we won
        win_tracker_yes_change.append(1)    # add 1 if we won
    else:
        win_tracker_yes_change.append(0)    # add 0 if we lost        

print('\n\n', 'WINNING PROBABILITY: ', sum(win_tracker_yes_change)/len(win_tracker_yes_change), '\n\n')
viz_func(win_tracker_yes_change)
```

#### Monty Haul with real probs

```python
movie_list = ['bad', 'middle', 'good']
choice_list = [0,1,2]

win_tracker_yes_change=[]

for i in range(1,10000):
    random.shuffle(movie_list)
    my_choice=random.choice(choice_list)
    
    ml= movie_list.copy()
    
    del ml[my_choice]
    
    if ml[0]!="good":
        del ml[0]
    else:
        del ml[1]
        
    if 'good'== ml[0]:
        win_tracker_yes_change.append(1)
    else:
        win_tracker_yes_change.append(0)
        
print(sum(win_tracker_yes_change)/len(win_tracker_yes_change))

#viz_func(win_tracker_yes_change)
```
