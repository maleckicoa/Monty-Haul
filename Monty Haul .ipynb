#### Viz Function

```python
def viz_func(win_tracker):
    imag=[]
    goat='/home/aleksa/Desktop/goat.png'
    money='/home/aleksa/Desktop/cars.png'

    for i in range(0,60): 
        if win_tracker[i]==0:
            imag.append(goat)
        else:
            imag.append(money)

    %matplotlib inline

    images = []
    for img_path in imag:#glob.glob('/home/aleksa/Desktop/*.png'):
        images.append(mpimg.imread(img_path))

    plt.figure(figsize=(40,10))# 
    columns = 20
    for i, image in enumerate(images):  
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        
        plt.imshow(image)
        plt.axis('off')
               
```

#### No Change

```python
movie_list = ['Goat', 'Goat', 'Money']
choice_list = [0,1,2]

win_tracker_no_change=[]

for i in range(1,1000):
    random.shuffle(movie_list)
    if 'Money'== movie_list[random.choice(choice_list)]:
        #print ('you won!')
        win_tracker_no_change.append(1)       
    else:
        #print('better luck next time')
        win_tracker_no_change.append(0)

print(sum(win_tracker_no_change)/len(win_tracker_no_change))

viz_func(win_tracker_no_change)
```

#### Change

```python


movie_list = ['Goat', 'Goat', 'Money']
choice_list = [0,1,2]

win_tracker_yes_change=[]

for i in range(1,10000):
    random.shuffle(movie_list)
    my_choice=random.choice(choice_list)
    
    ml= movie_list.copy()
    
    del ml[my_choice]
    
    if ml[0]=="Goat":
        del ml[0]
    else:
        del ml[1]
        
    if 'Money'== ml[0]:
        #print ('you won!')
        win_tracker_yes_change.append(1)
    else:
        win_tracker_yes_change.append(0)
        
print(sum(win_tracker_yes_change)/len(win_tracker_yes_change))

viz_func(win_tracker_yes_change)
```

#### Monty Haul with real probs

```python
movie_list = ['bad', 'middle', 'good']
choice_list = [0,1,2]

win_tracker_yes_change=[]

for i in range(1,10000):
    random.shuffle(movie_list)
    my_choice=random.choice(choice_list)
    
    ml= movie_list.copy()
    
    del ml[my_choice]
    
    if ml[0]!="good":
        del ml[0]
    else:
        del ml[1]
        
    if 'good'== ml[0]:
        win_tracker_yes_change.append(1)
    else:
        win_tracker_yes_change.append(0)
        
print(sum(win_tracker_yes_change)/len(win_tracker_yes_change))

#viz_func(win_tracker_yes_change)
```
